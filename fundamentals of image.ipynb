```python
import cv2
image = cv2.imread('./images/goldengate.jpg')

print(image)
```

```python
#show an image
cv2.imshow('output', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#color conversion

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('output', gray)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Create a blank image
import numpy as np

image = np.zeros((100, 100)) # 100 rows, 100 cols
cv2.imshow('output', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Checking the value of each pixel

for row in range(50):
    for col in range(50):
        print("Row : ",row,"Col :", col)
        print(image[row][col])
```

```python
#changing the value of each pixel
import random

#help(random.randint)

for row in range(100):
    for col in range(100):
        print("Row : ",row,"Col :", col)
        image[row][col] = int(random.randint(100,240))
        print(image[row][col])
```

```python
cv2.imshow('output', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cv2.imwrite('output.png', image)
```
